```python
# subset data frame to just species and year variables
df_subset = df[['species', 'year']]
# preview first 10 rows of new data frame
df_subset.head(10)
```

```python
# conditional subset; only rows where island is equal to Biscoe
df_subset = df[df['island'] == 'Biscoe']
# preview first 10 rows of new overwritten island subset
df_subset.head(10)
```

```python
# get a data frame of unique species, islands, and years
# drop duplicates and only include three variables
df_subset = df[['species','island','year']].drop_duplicates()
df_subset.head()
```

```python
# function order matters!
# drop duplicates only in these three variables but keep all variables
df_subset = df.drop_duplicates(subset = ['species','island','year'])
df_subset.head()
```

```python
# create new composite variable combining bill length and flipper length
df['bill_flipper_mm'] = df['bill_length_mm'] + df['flipper_length_mm']
df.head()
```

# Missing Data

```python
# we have some NaNs in new bill_flipper_mm variable
df_subset = df['bill_flipper_mm']
print(df_subset)
```

```python
# get row index of missing values
df_subset[df_subset.isna()].index
```

```python
# what other variables have missing data at these rows
df.iloc[[3,271]].isna()
```

```python
# store the row indicces of all missing values 
missing_indices = df[df.isna().any(axis=1)].index
# print
missing_indices
```

```python
# what variables have msising data in these rows
df.iloc[missing_indices].isna()
```

```python
# remove missing data in a new data frame
df_no_na = df.drop(index=missing_indices)
```

```python
# we can do this all in one step with a different function
df_no_na = df.dropna()
```

```python
# now all variables have equal number of non-null values
df_no_na.info()
```

# If / Else Statements

```python
# set x
x = 10
# if / else
if x > 5:
    print("x is less than 20")
else:
    print("x is greater than 20")
```

```python
# if / else with multiple conditions
if x < 5:
    print("x is less than 5")
elif x < 20:
    print("x is somewhere between 5 and 20")
else: 
    print("x is greater than 20")
```

```python
# get median value
median_value = df_no_na['bill_flipper_mm'].median()
# show
median_value
```

```python
# group penguin as "small" if bill_flipper_mm is below the median and "big" if above the median
df_no_na['bill_flipper_group'] = df_no_na.apply(lambda row: 'small' if
                                                row['bill_flipper_mm'] <= median_value
                                                else 'big', axis = 1)
df_no_na.head()
```

# For Loops

```python
for index, row in df_no_na[0:5].iterrows():
    print(f"Index: {index}, Length: {row['bill_flipper_mm']}, Group: {row['bill_flipper_group']}")
```
